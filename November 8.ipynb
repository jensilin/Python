```python
class yummy:
    def __init__(self,a):
        self.a=a
    
    def __add__(self,other):
        return self.a+other.a
a=yummy([1,3])
b=yummy([4])
# c=add([4])

# print(a)
print(a+b)
# print(type(a))
```

It is a phenomenon of making the operators work on user defined class(objects) or data types 

Magic Methods or Dunder method:
        Any method which is starting and ending with "Double underscore (__)" are called as magic method .
        It is not neccessary to call the magic method, by default it will be invoked.
        Usually operators will have magic methods


Operator Name : Addition
Magic Method : __add__
Internal Operation : ab1.__add__(obj2)

```python
#So initially addition is not possible in set if i want to add using set i should do operator overloading on __add__ magic method

class dummy:
    def __init__(self,a):
        self.a=a
    def __add__(self,other):
        for i in self.a:
            other.a.add(i)
        return other.a

ob1=dummy({1,2,3})
ob2=dummy({4,3,6})
print(ob1+ob2)
```

Regular Expression :

        Exept (-) other special characters have to be used with backward slash(\)

Regular Expression
    [A-Z] To match one Upper-case
    [a-z] one Lower case
    [0-9] one digit
    [A-Za-z0-9] one alphanumeric 
    [A-Z] {m} 'm' is the length of upper case
    [A-Z] {m,n} between m-n length uppercase
    * : 0 to  n
    + : 1 to n
    ? : 0 or 1
    \b : set boundaries
    \ : To remove special behaviour of special character


        PAN Card

Multi-threading -
        It used to make run the multiple program at the same time.
        It will reduce the time of execution. (Noramlly program will run one task and it will run the next task) ----- To run the different task at same time we use Multi-threading

        Drawbacks :
            Synchronisation process (The two task will not be performed simultaneously)

```python
import threading as th 
import time
def series():
    for i in range(1,11):
        print(i)
        time.sleep(1)
        
def series1():
    for i in range(1,11):
        print(i)
        time.sleep(1)
        
t1=th.Thread(target=series)
t2=th.Thread(target=series1)
t1.start()
t2.start()
t1.join()
t2.join()
```

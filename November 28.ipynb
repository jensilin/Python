User-defined packages or modules:
        Using keywords like from import it is possible to access modules or packages

        From keyword 
            It is a keyword which is used to point to a required location

        import 
            It is keyword which is used to flex the functionality
                Multiple files in a single package:
                    import filename
                    filename.pname/mname(args)

                    from filename import pname/mname
                    pname/mname(args)
                
                Multiple Folders / packages
                    from packagename import filename
                    filename.pname/mname(args)

                    from packagename.filename import pname/mname 
                    pname/mname(args)

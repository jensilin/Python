Decorator 
        It is a function which is used to  add additional features to the existing function without modifying it.
        The decortor is used on the functions which have common pre and post task.
        
        It is of two types:
            Inbuilt decorator
            User-defined decorator

        Inbuilt decorator : 
                These are pre-defined decorators to perform some task
                example: @classmethod @staticmethod @abstractmethod

        User-defined decorators
                These are the decorators which are created based on user requirements.
                Syntax: 
                    def decor_name(fname):
                        def inner(*args,**kwargs):
                            pre task
                            fname(*args,**kwargs):
                                post task
                            
                        return inner
                    
                    @decorator name
                    def fname(args):
                        S.B


        NOTE: A DECORATOR CAN HAVE ONLY PRE-TASK OR POST-TASK OR INCLUDING BOTH PRE-TASK AND POST-TASK

```python
def instagram(func):
    def inner(*args,**kwargs):
        print("Logged In")
        func(*args,**kwargs)
        print("logged out")
    return inner

@instagram
def Adhi():
    print("Seee reels")
    print("Seee story")
    print("Seee others")
    print("Seee see see")

Adhi()                  #using decorator we get logged in and logged out message
```

```python
@instagram
def renn():
    print("Hey hey hi")
    print("Hey i am using insta")
    print("So exited")
    print("Seeing reels")
    print(".......................")
    
    
renn()
```

```python
#WAP program which gives 5  sec before excuting any pogram using decorator

import time

def timely(hehe):       #3
    def inner(*args,**kwargs): #4 
        time.sleep(5) #6
        hehe(*args,**kwargs)  #7
        print("loo")  #9
    return inner    #5


@timely
def hehe():         #2
    print("Lo shyaaaaaaaa")   #8

hehe()              #1
```

```python
#WAP which calculates the total execution time of a program


def record(rec):
    def inner(*args,**kwargs):
        import time
        start=time.time()
        rec(*args,**kwargs)
        end=time.time()
        print(end-start)
    return inner

@record
def rec():
    for i in range(0,100):
        print(i)
        
rec()
```

```python
#Create a decorator which return positive values when an arthmethic operation is performed

def return_positive(arith):
    def inner(*args,**kwargs):
        res=arith(*args,**kwargs)
        print(res)
    return inner

@return_positive
def add(a,b):
    print(a+b)

add(-10,5)
```

Encapsulation / Access Specifier 
        Encapsulation is the phenomenon of providing security to the data 
        Access Specifiers are the members of class which specifies whether it can be accessed outside the class or not 

        It is of three types:
            Public
            Private
            Protected

        Public Access Specifier : These are the members of class which can be accessed outside the class 
            NOTE: ALL THE EXAMPLES SOLVED PREVIOUSLY ARE PUBLIC ACCESS SPECIFIER
            class cname:
                var=value
                def method_name(self,args):
                    S.B
                @classmethod
                def method_name(cls,args):
                    S.B

        Protected Access Specifier : These are the members of class which should provide security, but in python it does not provide any security.
            NOTE: WHEN A PROTECTED ACCESS SPECIFIER IS PRESENT IN A CLASS, AS A PROGRAMMER IT IS A CONVENTION THAT THESE MEMBERS SHOULD NOT BE USED IN SUB-CLASSES OR DIFFERENT FILE
            NOTE:A SINGLE UNDERSCORE HAVE TO BE SPECIFIED BEFORE THE VARIBLE OR METHOD DECLARATION 
            class cname:
                _var=value
                def _method_name(self,args):
                    S.B
                @classmethod
                def _method_name(cls,args):
                    S.B
            
        Private Access Specifier : These are the members of class which should provide security for the data and it cannot be accesed outside the class.
            NOTE:A SINGLE UNDERSCORE HAVE TO BE SPECIFIED BEFORE THE VARIBLE OR METHOD DECLARATION 
            class cname:
                __var=value
                def __method_name(self,args):
                    S.B
                @classmethod
                def __method_name(cls,args):
                    S.B

```python
#Public Access Specifier

class A:
    a=10
    b=20
    
ob=A()
print(ob.a,ob.b)
```

```python
#Protected Access Specifier

class C:
    _a=10
    _b=50
    c=28
hey=C()
print(hey._a,hey._b,hey.c) 
```

```python
#Private Access Specifier
class F:
    __a=50
    b=14
lo=F()
# print(lo.__a)                 Error will come because it is private
print(lo.b)                     #it is public
```

Syntax to access the private access specifier
    cname/obj._cname__pname
    cname/obj._cname__methodname(args)

```python
class Company:
    cname="Animation lab"
    watchman="gotilla"
    __turnover=700000
    
    def __init__(self,name,eid,sal):
        self.name=name
        self.eid=eid
        self.__sal=sal
        
    def __dis(self):
        print(self.name,self.eid,self.__sal)
    

obj=Company("jenny",5525,656456)

print(Company._Company__turnover)

print(obj._Company__dis())

print(obj._Company__sal)
        
```

Polymorphism 
        It is a phenomenon where a function or operator has more than one operation

        In python len function can be called as polymorphic because:
                for individual throws error
                In case of string it counts the no. of characters 
                In list tuple set it counts the number of values
                In dictionary it count keys

                Operators like + - * < >   etc.................... are considered as poymorphic operators

        

```python
len("deerftg")
```

```python
len([1,2,3,4])
```

```python
len((1,2,3,4,5))
```

```python
len({1,2,3,4,78,6,8})
```

```python
len({"desfgty":"wsadesfrgthy"})
```

Polymorphism can be explained used two scenarios

*************************************************************************************************************************************************************************************************
        Method Overloading:
            When two or more function having same name which works based on the number of arguments is called as method overloading
            NOTE: IN PYTHON METHOD OVERLOADING IS NOT POSSIBLE IT BECOMES OVERRIDING 
*************************************************************************************************************************************************************************************************

        Method Overriding:
            When two or more methods having same names is present it always takes a latest value by overriding the previous value.

        Operator Overloading:

Parsing Technique
        It is a phenomenon of providing security to the data while sending from one place to another (one file to another file) (source to destination)

        Two types
            JSON parsing technique
            Pickle parsing technique

        JSON parsing technique (Javascript object notation):
            It is a phenomenon of providing security to the data while sending from sourse to destination.
            In this case it will convert the original data in the form of string type (JSON Format)

            At destination point it will convert the string type into original data

            Encrypted data 
                import JSON
                var=JSON.dumps(ord_data)
            
            Decrypted data
                import JSON
                var=JSON.loads(enc_data)
            

```python
a=[10,20,30]
b=str(a)
print(type(b),b)
```

```python
org_data=[10,20,30]
import json
str_data=json.dumps(org_data)                   #Encrypt the data
print(org_data,type(org_data))
print(str_data,type(str_data))
```

```python
bb=json.loads(str_data)             #Decrypt the data
print(bb,type(bb))
```

Pickle parsing technique
        It is a phenomenon of providing security for the data while sending from source to destination in this case .
        In this case it will convert the data into binary form. 

        Syntax:
            Decripted data
                import pickle
                var=pickle_dumps(str_data)

            Decripted data
                import pickle
                var=pickle.loads(bin_data)

```python
ori_data=[30,40,50,60]
import pickle
bin_data=pickle.dumps(ori_data)
print(ori_data,type(ori_data))
print(bin_data,type(bin_data))
```

```python
decripy=pickle.loads(bin_data)
print(decripy,type(decripy))
```
